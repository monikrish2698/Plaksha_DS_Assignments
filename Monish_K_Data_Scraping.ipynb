```python
# importing the required libraries
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

### I have used selenium with beautiful soup. Wanted to get an exposure on how selenium can be used for scraping. Although, not used with in the best way but realised the importance of selenium which can help in interacting with web elements.

```python
# selenium creates a chrome instance through which it extracts the page source and the start scraping
driver = webdriver.Chrome("D:/chromedriver/chromedriver_win32/chromedriver.exe")
```

```python
full_df = pd.DataFrame(columns=["day", "temp", "short_desc", "desc"])
```

```python
link = "https://weather.com/en-IN/weather/tenday/l/d253814420804f28d49f7a75130e71b797c3c93f9ded842eaf6769e3784da352"
driver.get(link)
```

```python
sp = BeautifulSoup(driver.page_source, 'html5lib')
```

```python
full_df["day"] = [ele.text.strip() for ele in sp.find_all('h3', class_="DailyContent--daypartName--3emSU")]
full_df["temp"] = [ele.text.strip() for ele in sp.find_all('span', class_="DailyContent--temp--1s3a7")]
full_df["short_desc"] = [ele.text.strip() for ele in sp.find_all('div', class_="DailyContent--Condition--1zRBJ")]
full_df["desc"] = [ele.text.strip() for ele in sp.find_all('p', class_="DailyContent--narrative--3Ti6_")]
```

```python
driver.close()
```
